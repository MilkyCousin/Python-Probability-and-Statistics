```python
import matplotlib.pyplot as p
import numpy as n
import seaborn as s
```

```python
n.random.seed(0)
```

```python
# plot S_n/n, S_n := x_1 + x_2 + ... + x_n ; e - expected value of distribution of x_1 (x_j - i.i.d. for all 1 <= j <= n)
def g(x, y, e):
    f, a = p.subplots(1, 1, figsize=(12, 4))
    a.grid()
    s.lineplot(x, y, ax=a)
    s.lineplot(x, n.full(len(x), e), ax=a, color='r')
```

```python
N = 1000
q = 0.5
x_axis = n.arange(1, N+1)
```

```python
# based on geometric distribution
y_axis = n.cumsum(n.random.geometric(q, len(x_axis)))/x_axis
```

```python
g(x_axis, y_axis, (1/q))
```

```python
# based on discrete uniform distribution
y_axis = n.cumsum(n.random.randint(1, 6+1, len(x_axis)))/x_axis
```

```python
g(x_axis, y_axis, (1+2+3+4+5+6)/6)
```
